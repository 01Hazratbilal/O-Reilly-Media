There are a number of color abbreviations provided for commonly-used colors, but any color on the spectrum can be used by specifying its RGB value (for example, '#CECECE'). You can see the full set of linestyles by looking at the docstring for plot.

Line plots can additionally have markers to highlight the actual data points. Since matplotlib creates a continuous line plot, interpolating between points, it can occasionally be unclear where the points lie. The marker can be part of the style string, which must have color followed by marker type and line style

```python
plt.plot(np.random.randn(30).cumsum(), 'ko--')
```

This could also have been written more explicitly as:

```python
plt.plot(np.random.randn(30).cumsum(), color = 'black',
        linestyle = 'dashed', marker = '8')
```

For line plots, you will notice that subsequent points are linearly interpolated by default. This can be altered with the drawstyle option:

```python
data = np.random.randn(30).cumsum()
```

```python
plt.plot(data, linestyle = '--', marker = 'o', label = 'Default', color = 'blue')
```

```python
plt.plot(data, 'k-', drawstyle='steps-post', label='steps-post')
```

```python
plt.plot(data, marker = 'o', color = 'black',
        linestyle = '--', drawstyle = 'steps-post', label = 'steps')


plt.legend(loc = 'best')
```
