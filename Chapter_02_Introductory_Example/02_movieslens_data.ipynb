## MovieLens 1M Data set

#### ***GroupLens Research (grouplens.org)*** provides a number of collections of movies rating data collected from users of MoviesLens in the late 1990s and early 2000s.

#### The following data provide movies rating, movie metadata (gnres and year), and demographic data about the users (age, zip code, and occupation). Such data is often of interest in the development of recommendation systems based on machine learning algorithms.

> ### I will show how to slice and dice data sets like movielens data set into the exact form you need.

#### This data set contains 1 million rating s collected from 6000 users on 4000 movies. And spread across 3 tables:
1) ratings
2) User information
3) Movies Information

##### Each table can be loaded into a pandas DataFrame ofject using pandas.read_table:

```python
# import pandas to read table

import pandas as pd

user_name = ['user_id', 'gender', 'age', 'occupation', 'zip']
user = pd.read_table('../../CSV Files/User_Table(moviesdataset).dat', sep= '::', header = None, names= user_name, encoding= 'latin-1')

rating_names = ['user_id', 'movie_id', 'rating', 'timestanp']
rating = pd.read_table('../../CSV Files/Rating_Table(moviesdataset).dat', sep= '::', header = None, names= rating_names, encoding= 'latin-1')

movies_name = ['movie_id', 'title', 'genres']
movies = pd.read_table('../../CSV Files/Movies_Table(moviesdataset).dat', sep= '::', header = None, names= movies_name, encoding= 'latin-1')
```

> If you want to join 2 or more tables. You can use merge fuction to merge the tables and overlap the same rows (like user_id coulmn found in 2 of the tables, the merge fuction overlaf these values)

```python
# merge 2 and then merge the 3rd one and then 4th and so on

dat = pd.merge(rating, user)
data = pd.merge(dat, movies)
```

```python
data.iloc[0]
```

> Aggregating(کل  تعداد) the ratings grouped by one or more user of movie attributes is straightforward(واضع) once you build some familiarity with pandas. To get mean movies ratings for each film grouped by gender, we can use pivot_table method

```python
mean_rating = data.pivot_table('rating', index=  'title', columns= 'gender', aggfunc= 'mean')

mean_rating[:5]
```

##### This DataFame (mean_rating) containing mean ratings with movie totals as row labels and gender as column labels.

> Now, going to filter down to movies that recive at least 250 ratings (star) (a completely arbitrary number); to do this, I group the data by title and use size() to get a series of group size for each title:

```python
rating_by_title = data.groupby('title')

# rating_by_title = data.goroupby('title').size()
# Both have same output'

rating_by_title = rating_by_title.size()
```

```python
rating_by_title.head(10)
```

```python
active_titles = rating_by_title.index[rating_by_title >=250]
```

```python
active_titles[0:20]
```

##### The index of titles receiving at last 250 raings can then be used to select rows from mean_ratings above:

```python
mean_rating = mean_rating.loc[active_titles]

mean_rating
```

> To see the top films among female or male viewers, we can sort by the F or M column in descending or accending order:

```python
top_female_rating = mean_rating.sort_values(by = 'F', ascending = False ) 
# ascending False meand in decending order, and True means in ascending order

top_female_rating[:10]
```

```python
top_male_rating = mean_rating.sort_values(by = 'M', ascending = True)

top_male_rating[:5]
```

## Measuring rating disagreement

##### Suppose you wanted to find the movies that are most divisive between male and female viewers. One Way is to add a column to mean_ratings containing the difference in means, then sort by that:

```python
mean_rating['diff'] = mean_rating['M']- mean_rating['F']
```

```python
sort_by_diff = mean_rating.sort_values(by = 'diff', ascending= False)

sort_by_diff[:10]
```

##### Reversing the order of the rows and again sliceing off, we get the movies preferred by mean but women didn't rate as highly:

```python
# Reverse order of rows, take first 15 rows

sort_by_diff[::-1]
```

##### Suppose instead you wanted the movies that elicited the most disagreementamong viewers, independent of gender. Disagreement can be measured by the variance or standard deviation of the ratings:

```python
# Standard deviation of rating grouped by title
rating_std_by_title = data.groupby('title')['rating'].std()
```

```python
# Filter down to active_titles
rating_std_by_title = rating_std_by_title.loc[active_titles]
```

```python
# sort Series by value in descending order

rating_std_by_title.sort_values(ascending = False)

rating_std_by_title[:10]
```
